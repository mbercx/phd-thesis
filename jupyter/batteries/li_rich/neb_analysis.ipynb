```python
import os
import matplotlib.pyplot as plt
%matplotlib inline

import tikzplotlib
from pybat.core import DimerNEBAnalysis

figures_dir = "../../../figures/batteries"

color0 = "#163D5C"
color1 = "#B1CCE3"
color2 = "#3AA99F"
color3 = "#EECF5A"
color4 = "#E65239"
color5 = "#B81F1F"
color6 = "#2368A2"
color7 = "#0B164D"
color8 = "#0A5085"

SMALL_SIZE = 12
MEDIUM_SIZE = 14
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
neb = DimerNEBAnalysis.from_file("data/neb_Li2MnO3-49_80.json")
neb.remove_image(-2); neb.remove_image(-2); 
neb.setup_interpolation({"spline": "zero_slope"})
# plt.subplots_adjust(wspace=0.05, hspace=0.05)
# plt.savefig(os.path.join(figures_dir, "substitution_dimers.png"), dpi=300, 
#             bbox_inches="tight")
```

```python
fig, ax = plt.subplots()
neb.add_plot_to_axis(ax)

tikzplotlib.save(filepath=os.path.join(figures_dir, "neb_Li2MnO3-49_80-data.tex"), 
                 add_axis_environment=False)
```
