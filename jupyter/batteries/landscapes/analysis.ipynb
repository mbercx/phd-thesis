# Analysis of chain landscapes

In this notebook we'll test out our analysis scripts for the chained wedge landscapes, and use them to create the figures.

```python
# Notebook header

import os
import numpy as np
%matplotlib inline

from cage.landscape import LandscapeAnalyzer
from cage.cli.commands.analyze import sphere_analysis, landscape_analysis, \
    plot_landscape, distance

from support_functions import plot_min_energy_profile

data_dir = "/Users/mbercx/Dropbox/Phd/projects/llnl/data/"

def get_average_energy(landscape):
    return np.mean([d["energies"][-1] for d in landscape.data])
```

## CBH$_{11}$H$_{12}$

### Li

```python
# Load the data

ls_cb11h12 = landscape_analysis(
    directory=os.path.join(data_dir, "cb11h12", "Li", "chain_energy"),
    interp_method=("bisplrep", {"s": 0.1}),
    interp_mesh=(0.01, 0.01), 
    end_radii=(3.0, 4.4)
)
```

#### Plotting the landscapes

```python
plot_landscape(
    landscape_data=ls_cb11h12, 
    reference_energy=get_average_energy(LandscapeAnalyzer.from_file(
        os.path.join(data_dir, "cb11h12", "Li", "sphere_8.0", "landscape.json")
    )),
    energy_range=(-3, 0),
    set_contour_levels_manually=False, 
    colormap="RdYlBu_r",
)
```
